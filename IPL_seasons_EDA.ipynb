```python

year = 2008
win_per_season_ipl = pd.DataFrame(columns=['year', 'team', 'wins'])
for items in teams_per_season.items():
    if items[0][0] == year:
        print(items)
        win_series = pd.DataFrame({
            'year': [items[0][0]],
            'team': [items[0][1]],
            'wins': [items[1]]
        })
        win_per_season_ipl = pd.concat([win_per_season_ipl, win_series], ignore_index=True)
        year += 1
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from matches.csv
data = pd.read_csv('matches.csv')

# Filter the records
batting_first_win = data[data['win_by_runs'] > 0]
batting_second_win = data[data['win_by_wickets'] > 0]

# Count the number of wins for each team
batting_first_win_counts = batting_first_win['winner'].value_counts()
batting_second_win_counts = batting_second_win['winner'].value_counts()

# Combine the counts into a single DataFrame
team_win_counts = pd.DataFrame({
    'Team': batting_first_win_counts.index,
    'Batting First Wins': batting_first_win_counts.values,
    'Batting Second Wins': batting_second_win_counts.reindex(batting_first_win_counts.index, fill_value=0).values
})

# Plotting
team_win_counts.plot(x='Team', kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Number of Wins by Batting First and Batting Second')
plt.legend(title='Winning Scenario', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

import pandas as pd

# Load the data from matches.csv
data = pd.read_csv('matches.csv')

# Calculate the total number of matches
total_matches = len(ipl)

# Calculate the number of wins after winning or losing the toss for each team
win_after_winning_toss = ipl[ipl['toss_winner'] == ipl['winner']]
win_after_losing_toss = ipl[ipl['toss_winner'] != ipl['winner']]
win_after_winning_toss_counts = win_after_winning_toss['winner'].value_counts()
win_after_losing_toss_counts = win_after_losing_toss['winner'].value_counts()

# Calculate the win percentage after winning or losing the toss for each team
win_after_winning_toss_percentage = (win_after_winning_toss_counts / total_matches) * 100
win_after_losing_toss_percentage = (win_after_losing_toss_counts / total_matches) * 100

# Combine the counts and percentages into a single DataFrame
results = pd.DataFrame({
    'Wins after Winning Toss': win_after_winning_toss_counts,
    'Wins after Losing Toss': win_after_losing_toss_counts,
    'Total Matches': win_after_winning_toss_counts + win_after_losing_toss_counts,
    'Win Percentage after Winning Toss': win_after_winning_toss_percentage,
    'Win Percentage after Losing Toss': win_after_losing_toss_percentage
}).fillna(0)  # Fill NaN values with 0

# Display the results
print(results)
```

```python
# Plotting
results[['Wins after Winning Toss', 'Wins after Losing Toss']].plot(kind='bar', figsize=(12, 6))
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Number of Wins after Winning or Losing the Toss for Each Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
